# Regression with neural networks

```python
import pandas as pd
import numpy as np
df = pd.read_csv('filteredDatasets/740-NTH_elektriciteit_withWeather.csv')
df['Momentum'] = df['Momentum'].astype('datetime64')
df = df.set_index('Momentum', drop=False)
df.head()
```

The hour is coded by means of its sine and cosine as reported in <a href="http://adsabs.harvard.edu/abs/2007JEI....16d9901B">literature</a> and <a href="http://www.sciencedirect.com/science/article/pii/S0378778804003032">here</a>. It will be the same for the weekDay

```python
df['hour'] = 0
df['sinHour'] = 0
df['cosHour'] = 0
df['sinWeekDay'] = 0
df['cosWeekDay'] = 0
def sinDayAndHour(s):
    s['hour'] = s['Momentum'].hour
    return s
    
df = df.apply(sinDayAndHour, 1)
df['sinHour'] = np.sin(2*np.pi*df['hour']/23)
df['sinWeekDay'] = np.sin(2*np.pi*df['weekDay']/4)
df['cosHour'] = np.cos(2*np.pi*df['hour']/23)
df['cosWeekDay'] = np.cos(2*np.pi*df['weekDay']/4)
df.head()
```

In order to train the neural network easily, the weather columns will be normalized

```python
df['T_noNorm'] = df['T']
toNormalize = ['elektriciteit [kwh]', 'T', 'FH', 'U', 'N', 'Q']
for column in toNormalize:
    df[column] = (df[column] - df[column].mean()) / (df[column].max() - df[column].min())

df.head()
```

Since the days are very different from eachother, the dataset will be split in 3 datasets, and 3 different Neural networks will work on them.<br />
$NN_1=weekDays$<br />
$NN_2=saturdays$<br />
$NN_3=Sundays and holidays$<br /><br />

We will start from <b>the weekdays</b><br />

The features of the Neural networks are 10:<br />$x_1=elektriciteit [kwh]$<br />$x_2=T$<br />$x_3=FH$<br />$x_4=U$<br />$x_5=N$<br />$x_6=Q$<br />$x_7=sin(\pi*hour/12)$<br />$x_8=cos(\pi*hour/12)$<br />$x_9=sin(\pi*weekDay/2.5)$<br />$x_(10)=cos(\pi*weekDay/2.5)$ <br /><br />We are going to find the variable $y_0=gas [m3]$ thanks to a FeedFoward Neural network with Backpropagation. The tool used is pybrain.

```python
df = df[df.weekDay < 5]
```

```python
features = ['gas [m3]', 'elektriciteit [kwh]', 'T', 'FH', 'U', 'N', 'Q', 'sinHour', 'cosHour', 'sinWeekDay', 'sinWeekDay']
df[features].to_csv('filteredDatasets/740-NTH_NN_10F.csv', index=False)
```

## NN: previous gas consumption

Hypothesis: the new gas consumption depends also on the actual one.

```python
df['pre'] = df['gas [m3]'].shift(1)
df['pre'] = (df['pre'] - df['pre'].mean()) / (df['pre'].max() - df['pre'].min())
df['pre'].ix[0] = df['pre'].mean()
df.head()
```

```python
features = ['gas [m3]', 'elektriciteit [kwh]', 'T', 'FH', 'U', 'N', 'Q', 'sinHour', 'cosHour', 'sinWeekDay', 'sinWeekDay', 'pre']
df[features].to_csv('filteredDatasets/740-NTH_NN_11F.csv', index=False)
```

|---|---|---|---|---|---|
| 60 | 20 | 39.56 | 14.39 | 9.15 |
| 80 | 20 | 54.26 | 14.16 | 9.03 |

```python
rr = pd.read_csv('NNs/results/740-NTH_NN_11F.csv')
rr['actual'].head(80).plot()
rr['predicted'].head(80).plot()
```

Let's improve our hyphotesis with the dependency of the previous 2 hours

```python
df['pre2'] = df['gas [m3]'].shift(2)
df['pre2'] = (df['pre2'] - df['pre2'].mean()) / (df['pre2'].max() - df['pre2'].min())
df['pre2'].ix[0] = df['pre2'].mean()
df['pre2'].ix[1] = df['pre2'].mean()
df.head()
```

```python
features = ['gas [m3]', 'elektriciteit [kwh]', 'T', 'FH', 'U', 'N', 'Q', 'sinHour', 'cosHour', 'sinWeekDay', 'sinWeekDay', 'pre', 'pre2']
df[features].to_csv('filteredDatasets/740-NTH_NN_12F.csv', index=False)
```

## NN: previous gas consumption + Tdelta

An interesting feature can be the difference of temperature between now and the previous hour.

```python
df['Tdiff1'] = df['T_noNorm']-df['T_noNorm'].shift(1)
df['Tdiff1'] = (df['Tdiff1'] - df['Tdiff1'].mean()) / (df['Tdiff1'].max() - df['Tdiff1'].min())
df['Tdiff1'].ix[0] = df['Tdiff1'].mean()
df.head()
```

```python
features = ['gas [m3]', 'elektriciteit [kwh]', 'T', 'FH', 'U', 'N', 'Q', 'sinHour', 'cosHour', 'sinWeekDay', 'sinWeekDay', 'pre', 'pre2', 'Tdiff1']
df[features].to_csv('filteredDatasets/740-NTH_NN_13F.csv', index=False)
```

## NN: adding the temperature of the previous two days

```python
df['Tday-1'] = df['T_noNorm'].shift(1)
df['Tday-2'] = df['T_noNorm'].shift(2)
toNormalize = ['Tday-1', 'Tday-2']
for column in toNormalize:
    df[column] = (df[column] - df[column].mean()) / (df[column].max() - df[column].min())

    
df['Tday-1'].ix[0] = df['Tday-1'].mean()
df['Tday-2'].ix[0] = df['Tday-2'].mean()
df['Tday-2'].ix[1] = df['Tday-2'].mean()
df.head()
```

```python
features = ['gas [m3]', 'elektriciteit [kwh]', 'T', 'FH', 'U', 'N', 'Q', 'sinHour', 'cosHour', 'sinWeekDay', 'sinWeekDay', 'pre', 'pre2', 'Tdiff1', 'Tday-1', 'Tday-2']
df[features].to_csv('filteredDatasets/740-NTH_NN_15F.csv', index=False)
```
