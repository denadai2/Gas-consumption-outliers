# Regression with neural networks

Without a good model, the objective to detect outliers will not be achieved. It needs to be stated that the training of this model will be done from irregular data, containing also outliers; Neural networks are usually sensible to such defectives. Before explaining the Neural network, let's have a look on the data.

```python
import pandas as pd
import numpy as np
df = pd.read_csv('filteredDatasets/740-NTH_elektriciteit_withWeather.csv')
df['Momentum'] = df['Momentum'].astype('datetime64')
df = df.set_index('Momentum', drop=False)
df.head()
```

```python
df[df.weekDay < 5]['gas [m3]'].tail(80).plot()
```

Since the days are very different from eachother, the dataset will be split in 3 datasets, and 3 different Neural networks will work on them.<br />
$NN_1=weekDays$<br />
$NN_2=saturdays$<br />
$NN_3=Sundays and holidays$<br /><br />

We will start from <b>the weekdays</b><br />

The features of the Neural networks are 10:<br />$x_1=elektriciteit [kwh]$<br />$x_2=T$ (forecast)<br />$x_3=FH$ (forecast)<br />$x_4=U$ (forecast)<br />$x_5=N$ (forecast)<br />$x_6=Q$ (forecast)<br />$x_7=sin(\pi*hour/12)$<br />$x_8=cos(\pi*hour/12)$<br />$x_9=sin(\pi*weekDay/2.5)$<br />$x_(10)=cos(\pi*weekDay/2.5)$ <br /><br />We are going to find the variable $y_0=gas [m3]$ thanks to a FeedFoward Neural network with Backpropagation. The tool used is pybrain. <br />
We are going to minimize the RMSE and we will calculate also the MAE and MAPE.

The hour is coded by means of its sine and cosine as reported in <a href="http://adsabs.harvard.edu/abs/2007JEI....16d9901B">literature</a> and <a href="http://www.sciencedirect.com/science/article/pii/S0378778804003032">here</a>. It will be the same for the weekDay

```python
df = df[df.weekDay < 5]
```

```python
df['hour'] = 0
df['sinHour'] = 0
df['cosHour'] = 0
df['sinWeekDay'] = 0
df['cosWeekDay'] = 0
def sinDayAndHour(s):
    s['hour'] = s['Momentum'].hour
    return s
    
df = df.apply(sinDayAndHour, 1)
df['sinHour'] = np.sin(2*np.pi*df['hour']/23)
df['sinWeekDay'] = np.sin(2*np.pi*df['weekDay']/4)
df['cosHour'] = np.cos(2*np.pi*df['hour']/23)
df['cosWeekDay'] = np.cos(2*np.pi*df['weekDay']/4)
df.head()
```

In order to train the neural network easily, the weather columns will be normalized

```python
df['T_noNorm'] = df['T']
def normalizeColumns(df, columns):
    for column in toNormalize:
        df[column] = (df[column] - df[column].mean()) / (df[column].max() - df[column].min())
    return df

toNormalize = ['elektriciteit [kwh]', 'T', 'FH', 'U', 'N', 'Q']
df = normalizeColumns(df, toNormalize)
df.head()
```

```python
features = ['gas [m3]','elektriciteit [kwh]', 'T', 'FH', 'U', 'N', 'Q', 'sinHour', 'cosHour', 'sinWeekDay', 'cosWeekDay']
df[features].to_csv('filteredDatasets/740-NTH_NN_10F.csv', index=False)
```

## NN: adding exponential memory 

Always from [here](http://www.getacoder.com/data/projects/101950/Neural%20net%20architectures%20for%20temporal%20sequence%20processing.pdf), I add an exponential memory, which takes count of all the previous informaitons in an exponential way. I'll take count of electric consumption, Temperature and gas of the last 5 hours.

```python
def ci(t):
    u = 0.8
    return 1.0*(1-u)*(u**t)

def asdalolla(s):
    xr = 0.0
    for i in range(len(s)):
        xr = xr + ci(len(s)-i+1)*s[i]
    return xr

backtrack = ['elektriciteit [kwh]', 'T', 'gas [m3]']
for f in backtrack:
    temp[f+"back"] = pd.stats.moments.rolling_apply(temp[f],5,asdalolla)
toNormalize = ['elektriciteit [kwh]'+"back", 'T'+"back", 'gas [m3]'+"back"]
temp = normalizeColumns(temp, toNormalize)
    
features = ['gas [m3]', 'elektriciteit [kwh]', 'T', 'FH', 'sinHour', 'cosHour', 'sinWeekDay', 'cosWeekDay', 'gas1HBefore', 'gas2HBefore', 'Tdiff1', 'sinDayOfYear', 'cosDayOfYear','gasMean1DBefore','gasPeak1DBefore', 'gasSum1DBefore','gasPeak5HBefore','gasSum5HBefore','TPeak24HBefore','TSum24HBefore','gasPeak3HBefore','gasSum3HBefore', 'gasMean15DBefore', 'electrPeak3HBefore','electrSum3HBefore', 'gasPeak3DBefore','gasSum3DBefore','elektriciteit [kwh]'+"back", 'T'+"back", 'gas [m3]'+"back"]
temp[(temp['gas [m3]'].notnull()) & (temp['gasMean15DBefore'].notnull()) & (temp['gasPeak3DBefore'].notnull())][features].to_csv('filteredDatasets/prova9.csv', index=False)   
```
