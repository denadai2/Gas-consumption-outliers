# Outliers detection through GES

Let's try the <a href="http://www.itl.nist.gov/div898/software/dataplot/refman1/auxillar/esd.htm">extreme studentized deviate (ESD)</a>

```python
import csv
import numpy as np
import math
import matplotlib.pyplot as plt
#import pandas as pd
from PyAstronomy import pyasl
#
# Read dataset
#
f = open('nn1.csv', 'r')
data = csv.reader(f, delimiter=",")

nFeatures = 9
nOutput = 1
count = 0
GAS = []
for sample in data:
    #
    # Discard the label row
    #
    count = count +1
    if sample[0] == 'Momentum' or count > 5000000000000:
        continue

    label, x = sample[-1], [sample[1], sample[2], sample[3], sample[4], sample[5], sample[6], sample[7], sample[8], sample[9]]
    
    #
    # Insert the sample into the Dataset
    #
    GAS.append(label)
```

```python
GAS = np.array(map(lambda x: float(x), GAS))
r = pyasl.generalizedESD(GAS, 10, 0.05, fullOutput=True)
```

```python
print "Number of outliers: ", r[0]
print "Indices of outliers: ", r[1]
print "        R      Lambda"
for i in range(len(r[2])):
  print "%2d  %8.5f  %8.5f" % ((i+1), r[2][i], r[3][i])

figure(figsize=(15,7))
# Plot the "data"
plt.plot(GAS, 'b.')
# and mark the outliers.
for i in range(r[0]):
  plt.plot(r[1][i], GAS[r[1][i]], 'rp')
plt.show()
```
